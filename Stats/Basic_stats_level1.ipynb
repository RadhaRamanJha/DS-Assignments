## Q.no.3

```python
from scipy.stats import binom
# Number of coin tossed = 3
n = 3
# Probability of getting a head on a single coin toss
p = 0.5
# Number of heads you want
k = 2
# Calculate the probability using the binomial distribution
probability = binom.pmf(k, n, p)
print(f"The probability of getting {k} heads and {n-k} tails is: {probability:.4f}")
```

## Q.no.4 a)

## Equaul to one

```python
from itertools import product

# Define the sides of a die
sides_of_die = [1,2,3,4,5,6]

# Generate all possible outcomes of rolling two dice
all_outcomes = list(product(sides_of_die, repeat=2))

# Count the number of outcomes where the sum is 1
sum_is_1_outcomes = [(x, y) for x, y in all_outcomes if x + y == 1]

# Calculate the probability
probability = len(sum_is_1_outcomes) / len(all_outcomes)

print(f"The probability of getting a sum of 1 is: {probability:.4f}")
```

## b)	Less than or equal to 4 

```python
# Count the number of outcomes where the sum is less  than equal to 4
sum_le_or_eq_4_outcomes = [(x, y) for x, y in all_outcomes if x + y <= 4]

# Calculate the probability
probability = len(sum_le_or_eq_4_outcomes) / len(all_outcomes)

print(f"The probability of getting a sum less than equal to 4 is: {probability:.4f}")
```

## Sum Divisible by 2 and 3 

```python
# Count the number of outcomes where the sum is divisible by 2 and 3
sum_div_by_two_and_three_outcomes = [(x, y) for x, y in all_outcomes if ((x+y)%2 == 0 and (x+y)%3 == 0)]

# Calculate the probability
probability = len(sum_div_by_two_and_three_outcomes) / len(all_outcomes)

print(f"The probability of getting a sum divisible by 2 and 3 : {probability:.4f}")
```

## Q 5

```python
from math import comb

# Number of red, green, and blue balls
red_balls = 2
green_balls = 3
blue_balls = 2

# Total number of balls
total_balls = red_balls + green_balls + blue_balls

# Number of ways to choose 2 red balls from the non-blue balls
ways_to_choose_2_red = comb(red_balls + green_balls, 2)

# Total number of ways to choose 2 balls from the total
total_ways_to_choose_2 = comb(total_balls, 2)

# Probability of drawing two balls that are not blue
probability_none_blue = ways_to_choose_2_red / total_ways_to_choose_2

print(f"The probability that none of the balls drawn is blue is: {probability_none_blue:.4f}")
```

## Q.22- Calculating z-score for given confidence interval

```python
from scipy import stats
# Confidence intervals
cis = [.90,.94,.60]
# Z-scores
for ci in cis:
    z_score = stats.norm.ppf((1+ci)/2)
    print(f'z-score for {ci*100}% Confidence level is {z_score:.3f}')
```

## Q.23- Clacultating t-score for given confidence interval

```python
# Sample number
sample_num = 35
dof = sample_num-1
# Confidence intervals
cis = [0.95,0.96,0.99]
# t-score
for ci in cis:
    t_score = stats.t.ppf((1+ci)/2,df=dof)
    print(f'T-score for {ci*100} % Confidence level is {t_score:.3f}')
```

## Q 24 :- The probability that 18 randomly selected bulbs would have an average life of no more than 260 days

```python
# Given Values 
sample_mean = 260
population_mean = 270
std_dev = 90
num_sample = 18
t_score = (sample_mean-population_mean)/((std_dev)/np.sqrt(num_sample))
print(f"T-Score : {t_score}")
```

```python
dof = num_sample-1 # Degree of freedom
prob = stats.t.cdf(t_score,dof)
print(f"Probailty that 18 randomly selected bulbs would have an average life of no more than 260 days if CEO claim were true is {round(prob*100,2)} %")
```
